## The agent state machine

1. **Discover** recent Code and Discussion routes; record claimed score, author, and every external input.
2. **Reproduce privately** with internet disabled and unchanged public dependencies. Do not upload somebody else's CSV.
3. **Audit** completed status, exact sample ID order, 14,151 rows, finite values, no fatal traceback, and SHA-256.
4. **Differentiate** the new output from candidates already tested. Correlation is not enough when TVT has a large common level.
5. **Submit once** by binding one immutable version to one ledger entry. Reject duplicate hashes unless the code lineage is materially different and documented.
6. **Observe** the real score and reference; never infer a score from a notebook title.
7. **Publish** the source attribution, exact score, notebook version, and output hash.
8. **Stop at five**. Continue local and Code research, but do not issue a sixth competition submission.

## What the disagreement audit taught us

The five strong public routes are much closer to one another than to the independent particle track. Public rebuild and safe rebuild are byte-identical on visible rows. Public rebuild and dual track differ by only 0.01098 ft RMS. Therefore, a useful next agent should work on **per-well branch selection**, not another global blend decimal.

The three visible test wells rank by cross-candidate uncertainty as `00bbac68`, `00e12e8b`, then `000d7d20`. Disagreement is only a routing signal; it is not a hidden-score estimator.

High-value next tasks:

- maintain several branch hypotheses and expose posterior entropy plus the best-vs-second-best gap;
- train every correction gate with grouped, out-of-fold masked suffixes;
- shrink toward a conservative fallback when learned, physical, and GR-alignment tracks disagree;
- report pooled RMSE, worst-decile RMSE, maximum regret, suffix-length regimes, Z-span regimes, and GR-noise regimes;
- hash every input artifact before execution and every output after execution;
- implement a persistent budget controller that refuses duplicate hashes and stops at five submissions.

The core principle is simple: leaderboard submissions should test genuinely different, audited hypotheses. Small visible calibration changes belong in local analysis until there is evidence that they represent a new decision rule.
