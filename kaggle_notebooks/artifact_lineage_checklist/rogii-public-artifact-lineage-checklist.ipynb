The duplicate-hash check is deliberate. Public rebuild and safe rebuild are byte-identical on visible rows. A budget controller should reject an accidental duplicate automatically. If two code lineages with the same visible hash are intentionally tested, the ledger must say that explicitly.

## Pre-run artifact contract

An agent should save:

- source notebook URL, author, pulled version, and pull timestamp;
- every dataset/model slug and expected file pattern;
- hashes for critical model files when the platform exposes them;
- internet, accelerator, image, and package assumptions;
- a declared fallback path for every optional artifact.

A missing artifact must either stop the run or produce an explicit fallback manifest. Silent fallback makes score attribution invalid.

## Post-run submission contract

Before an agent is allowed to submit:

1. the private notebook status is `COMPLETE`;
2. `submission.csv` has exactly `id,tvt` columns and 14,151 rows;
3. IDs match the official sample in exact order and are unique;
4. every prediction is finite;
5. logs contain no fatal traceback;
6. SHA-256 is persisted before submission;
7. the daily ledger has capacity and has not seen the hash already;
8. one notebook version is bound to one submission reference;
9. the real score is attached only after Kaggle reports `COMPLETE`;
10. any public fork retains clear source attribution.

This workflow is useful beyond ROGII: it prevents agents from confusing a copied output, a silent fallback, and a genuinely reproduced model.
